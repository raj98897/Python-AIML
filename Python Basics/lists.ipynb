Practice Problem: Take a list and reverse the order of its elements.

```python
num_list = [100, 200, 300, 400, 500]
left = 0
right = len(num_list) - 1
while left < right:
    temp = num_list[left]
    num_list[left] = num_list[right]
    num_list[right] = temp
    left += 1
    right -= 1

print(num_list)            
```

```python
reversed_list = num_list[::-1]
print(f"Reversed list: ", reversed_list)
```

Practice Problem: Determine if a list reads the same forward and backward. The function should return True if it is a palindrome and False otherwise.

```python
numbers = [1, 2, 3, 4, 5, 4, 3, 2, 1]
first = 0
last = len(numbers) - 1
flag = False
while first < last:
    if numbers[first] == numbers[last]:
        first += 1
        last -= 1
        flag = True
    else: 
        flag = False  
        break 
if flag:
    print(True)
else:
    print(False)             
```

```python
def is_Palindrome(lst):
    return lst == lst[::-1]

test1 = [1, 2, 3, 2, 1]
test2 = [1, 2, 3, 4, 4]

print(is_Palindrome(test1))
print(is_Palindrome(test2))
```

Practice Problem: Create a function that determines if a list of numbers is sorted in non-decreasing (ascending) order. Return True if it is, and False otherwise.

```python
numbers = [10, 20, 30, 25, 40]
is_sorted = False
for i in range(len(numbers)):
    if numbers[i] < numbers[i + 1]:
        is_sorted = True
    else:
        is_sorted = False
        break
if is_sorted:
    print(True)
else:
    print(False)            
```

Practice Problem: Write a function to find the Power Set of a given list. The Power Set is a list of all possible subsets, including the empty list and the list itself.

```python
def power_set(nums):
    result = [[]]   # start with empty set
    for num in nums:
        new_subsets = []
        for subset in result:
            new_subsets.append(subset + [num])
        result.extend(new_subsets)
    return result

print(power_set([1,2]))

```
